# 畫函數圖形(上課練習)

畫$f(x)=\sin(x)$

方法一：定義ｘ＆ｙ，並加入圖表中的標籤、網格跟標題。

```python
# Generate x values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 100) # np.linspace(start, stop, num)，生成一个從start(起始值，此為0)到stop(終止值，此為2π)的等間距數列，num為要生成的點個數(此為100個點)。

# Calculate y values using the sine function
y = np.sin(x)

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Plot of y = sin(x)")

# Display the plot
plt.grid(True) # 顯示網格
plt.show()
```

畫$f(x)=\sin(x)$

方法二：僅通過簡單定義ｘ＆ｙ，也可以畫出函式圖形來！但就相對簡單一些。

```python
# Generate x values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 1000) # 此處將num(要生成的點個數)改成1000個點，看起來圖形一樣平滑!

# Calculate y values using the sine function
y = np.sin(x)

# Create the plot
plt.plot(x, y)
```

畫$f(x)=\sin(x)$

方法三：一樣是簡單定義ｘ＆ｙ，但調整參數linewidth，將函數寬度改成是 10 ，讓圖形加粗！

```python
# Generate x values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 1000)

# Calculate y values using the sine function
y = np.sin(x)

# Create the plot
plt.plot(x, y, linewidth=10)
```

# 額外作業練習

前言：數學函式千變萬化，有些函數可以做到許多厲害的應用；也有許多函數有充滿藝術感的氣息；更有些函數有永恆、無限、天長地久的額外意涵。

像是玫瑰線、心型線、螺旋、Weierstrass’ function等都是相當經典的圖形，其中莫比烏斯環更是令我驚艷，它是一種只有一個表面和一條邊界的曲面，由此也衍生出許多視覺藝術等領域的應用，而我也想藉此作業的機會，能更了解如何通過程式碼來畫出此奧妙且永恆的數學函數。

詳細內容如下：


```python
%matplotlib inline　            # 讓你可以内嵌繪圖，後續能省略掉plt.show()這段程式碼。

import numpy as np            # 針對科學運算開發的函式庫，提供多樣陣列功能與數學運算工具。
import matplotlib.pyplot as plt      # 繪圖函式庫，提供2D繪圖和3D繪圖套件。
from mpl_toolkits.mplot3d import Axes3D  # matplotlib的子庫，能更好創建3D座標並繪製3D座標、線跟面等。

# 定義參數範圍
ring_radius = -2       # 環半徑（影響整體大小）
strip_width = 2       # 環寬度（影響環的厚度）
colormap_style = 'plasma'  # 可選viridis(黃綠色系，適合高對比顯示。),coolwarm(藍到紅的過渡，適合表達溫度變化。),twilight(紫色粉色漸變，較夢幻。),plasma(紫、粉、黃，對比度高。)等顏色定義。
bg_color = '#97CBFF'   # 背景顏色
elev, azim, dist = 20, 30, 10  # 視角調整、elevation、azimuth、distance(仰角、方位角、距離)(詳細參考下方附錄二)
u = np.linspace(0, 2 * np.pi, 80)  # 角度（𝑢在區間[0,2𝜋]內，控制環的環繞程度。）(num_points = 80是看要等距分成幾格的參數，如此處就是產生一個從0~2π、等距切分成80點的環。)
v = np.linspace(ring_radius, strip_width, 50)  # 寬度（v在區間[ring_radius,strip_width]內，決定莫比烏斯帶的寬度。）
U, V = np.meshgrid(u, v)  # 建立2d網格來幫助理解3d圖形

# 莫比烏斯帶參數方程(詳細參考下方附錄一)
X = (1 + V * np.cos(U / 2)) * np.cos(U)
Y = (1 + V * np.cos(U / 2)) * np.sin(U)
Z = V * np.sin(U / 2)

# 設定圖形大小跟設定
fig = plt.figure(figsize=(8, 6)) # 繪製圖形外框的長跟寬
ax = fig.add_subplot(111, projection='3d', facecolor=bg_color) # 1 row、1 cols、1index，表示整張是唯一的3d圖，並將projection設為'3d'，建立3d坐標軸來繪製3d圖。

# 繪製圖形及顏色變化
color_map = np.sin(U)  # 使用 U 軸的正弦值作為顏色變化
ax.plot_surface(X, Y, Z, facecolors=plt.cm.get_cmap(colormap_style)(color_map), edgecolor='k', linewidth=0.3) # 繪製圖形

# 隱藏軸線、刻度線、標籤
ax.set_xticks([])  # 隱藏x軸數值跟格線
ax.set_yticks([])  # 隱藏y軸數值跟格線
ax.set_zticks([])  # 隱藏z軸數值跟格線
# ax.set_axis_off()  # 隱藏所有x、y、z軸
```
