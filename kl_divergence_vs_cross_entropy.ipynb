## 1. 解釋$cross$ $entropy$為何物?

起因：

希望在輸出結果都是正確的情況下，兩者計算loss的結果可以是一樣的!

解釋：

觀測預測的機率分佈與實際機率分布的誤差範圍，利用理論綜合實際資訊之散布程度，資訊量的大小可以看作是事件給我們的驚訝程度。

預測值與實際值差越多，越是不容易發生的事件，預測錯誤時，代表內涵的資訊量愈大，也就是不確定越多，$cross$ $entropy$會越高。

反之預測值與實際值差不多，預測正確時，$cross$ $entropy$會越低。

Cross entropy measures the average number of bits needed to encode events from one probability distribution (P) using the optimal code for another probability distribution (Q).

算式：

$H(P, Q) = - ∑ P(x) * log(Q(x))$

當有三個可能事件時，交叉熵的定義是： $H(P, Q) = -P1log Q1-P2log Q2-P3*log Q3$　！（類似期望值的算法）

目標：

 minimize the cross entropy between the predicted probability distribution and the true distribution.

搭配老師上課講解:

傳統計算Loss的方式通常是藉由去計算「預測值與真實值之間的距離平方和(MSE、均方誤差)」，先通過One-Hot Encoding進行編碼，打破數字間的連續性，變成是 N 維向量來表示是否為該特徵標籤，再藉由Softmax 函數把 N 維向量轉為是個機率分布的型態，總和等於 1 。

$但在正確答案為(1, 0, 0)時，明明(0.6, 0.3, 0.1)以及(0.6, 0.2, 0.2)的輸出結果都是正確的，可是在計算loss$時，兩者的差距卻是不一樣的$(0.26$  VS  $0.24)$，顯示出在明明輸出結果都一樣正確情況下，loss算出的結果卻有好壞之分，因此會希望在輸出結果都正確的情況下，可以讓loss的結果也能是一樣的！知道最終結果其實是一樣好的！$

計算「熵（entropy）」，也就是「平均訊息量」或是「平均驚訝指數」。
發生機率比較小的事，訊息量其實是比較大的。若要定義「訊息量」，應該是機率愈接近 0 ，訊息量愈大；而當機率是 1 的時候，訊息量是 0。即可定義為「 $I = -log P$ 」

所謂的「熵(entropy)」就是指「平均訊息量」或是「平均驚訝指數」。
假設有個機率分布 P，有三個可能的事件，機率分別是$ P1、P2、P3$。這時相對的訊息量就會是 $-log P1、-log P2、-log P3$。此時平均訊息量的機率分布期望值就會是：
$H(P) = -P1*log P1-P2*log P2-P3*log P3$，平均訊息量越大，我們越難知道資訊，也就是此機率分布的「亂度」比較大。

若正確答案是第 1 類時，我們希望 AI 輸出的 $-log Q1$ 愈小愈好。如果 AI 模型說出正確答案的機率很低，我們就要給它大扣分！
也因此交叉熵(cross entropy)在一般兩個機率分布 $P＝[P1,P2,P3]，Q＝[Q1,Q2,Q3]$ 時，交叉熵的定義是：
$H(P, Q) = -P1*log Q1-P2*log Q2-P3*log Q3$，若能證明 P 和 Q 的機率分布完全一樣，那麼交叉熵就是最小的 0 。

# $KL$ $divergence$

image.png

## 1. 解釋$KL$ $divergence$為何物?

起因：

$KL divergence（Kullback-Leibler Divergence），正確答案(真實答案、模型的訓練資料)寫成 P 機率分布，模型預測寫成 Q 機率分布，一樣是想辦法讓 Q 越來越像 P ，但若 P 不是 One-Hot Encoding 形式時，即使模型預測和正確答案一模一樣，loss function一樣也不會是 0 ！？$


概念與算式：


於是 $KL$ $divergence$ 借用$Cross$ $Entropy$的概念，讓模型預測跟真實機率(正確答案、模型的訓練資料)一樣時，誤差也能是 0，就是通過交叉熵減去正確答案熵的方式！$DKL(P∥Q)＝H(P, Q) - H(P)$，其實就是$Cross$ $Entropy$扣掉目標分布的$Entropy！$

計算機率分布，P&Q相似程度。P表示資料的真實分佈，Q表示資料的理論分佈，模型分佈，或P的近似分佈。可以視為分布Q用於近似P的訊息損失。
$為非對稱量度，D(P||Q)與D(Q||P)不相等，並不是兩個分布之間真正的距離。$

預測準確時，$KL$ $divergence$較小；預測錯誤時，$KL$ $divergence$數值變大。

KL divergence measures the difference between two probability distributions P and Q. It quantifies the amount of information loss when Q is used to approximate P.



---



---



搭配老師上課講解:

當有通過「One-Hot Encoding」進行編碼時，即可使用cross entropy的方式來繼續loss function，因為此時真實答案的機率分布，除了正解為 1 之外，其餘都是 0 ！因此只須關注模型預測出來的機率分布，再正解上是否機率也可以愈接近 1 ！若愈接近 1 則扣越少分數，機率越小則扣一大堆分數！！

因此遇到分類任務、只有一個正確答案，且計算loss function只在意正確答案情況時，使用cross entropy都相當適合！

－－－－－－

由於通常cross entropy的結果都不太可能等於 0 ，通常都是H(P)。
因此kl divergence出現了！通過DKL(P∥Q)＝H(P, Q) - H(P)，最小就可以輕易變成 0 了！此外他還有個嚇人的公式，歡迎查詢了解並寫入作業

－－

或是說：我們要比較兩個分布差多遠？也就是我們要讓Ｑ花多少能量變成Ｐ，花越多能量，代表他們差別比較大；花越少能量代表差別越小。而當Ｐ＝Ｑ時，kl divergence就會是 0 ！但通常cross entropy的結果都不太可能等於 0 。

－－－

有人會說cross entropy比較不會在意非真實答案（已知訓練資料）其他的值，而kl divergence會比較重視其他的，對其他進行懲罰的效果，但這不完全正確！

因為已知訓練資料中真實值的entropy（H(P)）是不會改變的！所以會差的只有前面的cross entropy（H(P, Q)）！！因此差的數字是一模一樣的！！

兩者差的數字是一樣的，主要只是差在比例的關係！！kl divergence可以更好看出兩者的好壞差距大小！

－－－－

我們希望能夠比較好看出差異的級距就使用kl divergence！

# 2. 舉例說明什麼情境用哪個?什麼情況能混用?

cross entropy使用時機：
1. 遇到分類任務、只有一個正確答案，且計算loss function只在意正確答案情況，目標是最小化預測機率和真實機率（即基本事實標籤）之間的差異時，使用cross entropy都相當適合！
2. 與機率分布、模型訓練有關的任務，用 Softmax 輸出的模型等。
3. Decision Trees, similarity metrics, dimension reduction algorithms等都是entropy可使用到的範圍!
4. 通常用於具有固定目標分佈的監督學習任務。


KL divergence使用時機：
1. 如果有兩個獨立的機率分布p(x)和q(x)同時對應到同一個隨機變數x，也就是它們所在的空間是一樣的，則可以使用KL Divergence來測量這兩個分布的差異程度。
2. 我們希望能夠比較好看出差異的級距就使用kl divergence！(比較兩者的差距)
3. 當我們使用「蒸餾」策略，需要學soft target時，就該考慮使用kl divergence！老師教學生，學生是想辦法學老師的輸出，不是學正確答案(1,0,0)，而是老師的(0.7,0.25,0.05)!![使用概率模型來近似另一個分布]
4. 也適用於聚類、降維和密度估計等任務上，其目標是透過最小化真實資料分佈和學習模型分佈之間的差異來揭示資料中的結構。
5. 通常用於非監督式學習任務和涉及兩個機率分佈的比較或近似的應用。

什麼情況可以混用:
1. 有通過「One-Hot Encoding」進行編碼時!兩者算出來的答案將是一樣的!

# 3. 寫程式來實驗上述情況!?（準備一些例子，來解釋兩者哪裡不一樣？）


```python
# 引入老師的固定4行套件
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import softmax
from scipy.stats import entropy
```

## 當 P 是 One-Hot 分布時，我們可以看出cross entropy = KL Divergence。因為此時 H(P) = 0

```python
# One-hot 分布
P = np.array([1.0, 0.0, 0.0])  # 真實標籤
Q = np.array([0.7, 0.2, 0.1])  # 預測機率分布

# 計算 KL Divergence
kl_div = entropy(P, Q)  # Scipy 的 entropy(p, q) 計算的是 D_KL(P || Q)

# 計算cross entropy
cross_entropy = -np.sum(P * np.log(Q))

print(f"KL Divergence: {kl_div:.4f}")
print(f"cross entropy: {cross_entropy:.4f}")
```

## 當 P 是機率分布時，cross entropy = KL Divergence + entropy(P)。

## 可以看出KL Divergence會比較小

```python
# 讓 P 是一個機率分布
P = np.array([0.5, 0.4, 0.1])  # 真實分布
Q = np.array([0.7, 0.2, 0.1])  # 預測機率分布

# 計算 KL Divergence
kl_div = entropy(P, Q)

# 計算cross entropy
cross_entropy = -np.sum(P * np.log(Q))

print(f"KL Divergence: {kl_div:.4f}")
print(f"cross entropy: {cross_entropy:.4f}")
```

## 當 P 沒通過softmax層時
## 可以看出使用cross entropy來計算loss function時，兩者差距又比機率分布時來得更大

```python
# 當 P 沒通過softmax層時
P = np.array([1.5, 1.2, 0.3])  # 真實機率分布*3
Q = np.array([0.7, 0.2, 0.1])  # 預測機率分布

# 計算 KL Divergence
kl_div = entropy(P, Q)

# 計算cross entropy
cross_entropy = -np.sum(P * np.log(Q))

print(f"KL Divergence: {kl_div:.4f}")
print(f"cross entropy: {cross_entropy:.4f}")
```

## 當 P 是 One-Hot 分布時，我們比較更多不同分布計算loss的結果，一樣可以看出cross entropy = KL Divergence。

```python
# 定義 One-hot 標籤
P = np.array([1.0, 0.0, 0.0])

# 定義三種不同的預測機率分布
Q_A = np.array([0.9, 0.05, 0.05])  # 接近真實答案
Q_B = np.array([0.6, 0.3, 0.1])   # 有點偏差
Q_C = np.array([0.2, 0.4, 0.4])   # 完全錯誤的預測

# 計算 KL Divergence
kl_A = entropy(P, Q_A)
kl_B = entropy(P, Q_B)
kl_C = entropy(P, Q_C)

# 計算cross entropy
cross_entropy_A = -np.sum(P * np.log(Q_A))
cross_entropy_B = -np.sum(P * np.log(Q_B))
cross_entropy_C = -np.sum(P * np.log(Q_C))

# 輸出結果
print(f"模型 A（接近真實標籤）")
print(f"  KL Divergence: {kl_A:.4f}, cross entropy: {cross_entropy_A:.4f}\n")

print(f"模型 B（有點偏差）")
print(f"  KL Divergence: {kl_B:.4f}, cross entropy: {cross_entropy_B:.4f}\n")

print(f"模型 C（預測錯誤）")
print(f"  KL Divergence: {kl_C:.4f}, cross entropy: {cross_entropy_C:.4f}\n")
```

## 當 P 是機率分布時，比較三種不同機率分布的狀況即可清楚發現，cross entropy跟KL Divergence兩者間差距真的都是一樣的!

但倆倆的差距就很明顯看得出來，通過計算兩兩不同模型的cross entropy或KL Divergence，

在cross entropy看起來可能差不多，只不過差了0.15的樣子，0.9725變成0.8286，好像也還好。

但KL Divergence，一樣只差0.15，但數字直接從0.1707變成是0.0268，一差就差了6.3647倍!很快就能看出兩者差異!

```python
# 定義機率分布標籤
P = np.array([0.7, 0.2, 0.1])

# 定義三種不同的預測機率分布
Q_A = np.array([0.9, 0.05, 0.05])  # 預測正確一
Q_B = np.array([0.6, 0.3, 0.1])   # 預測正確二
Q_C = np.array([0.2, 0.4, 0.4])   # 完全錯誤的預測

# 計算 KL Divergence
kl_A = entropy(P, Q_A)
kl_B = entropy(P, Q_B)
kl_C = entropy(P, Q_C)

# 計算cross entropy
cross_entropy_A = -np.sum(P * np.log(Q_A))
cross_entropy_B = -np.sum(P * np.log(Q_B))
cross_entropy_C = -np.sum(P * np.log(Q_C))

# 輸出結果
print(f"模型 A（預測正確一）")
print(f"  KL Divergence: {kl_A:.4f}, cross entropy: {cross_entropy_A:.4f}\n 兩者差距: {cross_entropy_A - kl_A:.4f}\n 兩者比值: {cross_entropy_A / kl_A:.4f}\n")

print(f"模型 B（預測正確二）")
print(f"  KL Divergence: {kl_B:.4f}, cross entropy: {cross_entropy_B:.4f}\n 兩者差距: {cross_entropy_B - kl_B:.4f}\n 兩者比值: {cross_entropy_B / kl_B:.4f}\n")

print(f"模型 C（預測錯誤）")
print(f"  KL Divergence: {kl_C:.4f}, cross entropy: {cross_entropy_C:.4f}\n 兩者差距: {cross_entropy_C - kl_C:.4f}\n 兩者比值: {cross_entropy_C / kl_C:.4f}\n")

print(f"模型 A 的 KL Divergence: {kl_A:.4f}, 模型 B 的 KL Divergence: {kl_B:.4f}\n 兩者比值: {kl_A / kl_B:.4f}\n")
print(f"模型 A 的 cross entropy: {cross_entropy_A:.4f}, 模型 B 的 cross entropy: {cross_entropy_B:.4f}\n 兩者比值: {cross_entropy_A / cross_entropy_B:.4f}\n")
```

## 當 P 沒通過softmax層時
## 可以看出在同模型間loss function差距更大
## 但在沒通過softmax的狀況下，兩者具有明顯區別， KL Divergence沒有任何改變，但cross entropy的loss值又差異的更大了!
## 同時也可清楚看出A & B兩種預測，B的預測結果相對比 A 更好！

```python
# 當 P 沒通過softmax層時
P = np.array([2.1, 0.6, 0.3])   # 真實機率分布*3

# 定義三種不同的預測機率分布
Q_A = np.array([0.9, 0.05, 0.05])  # 預測正確一
Q_B = np.array([0.6, 0.3, 0.1])   # 預測正確二
Q_C = np.array([0.2, 0.4, 0.4])   # 完全錯誤的預測

# 計算 KL Divergence
kl_A = entropy(P, Q_A)
kl_B = entropy(P, Q_B)
kl_C = entropy(P, Q_C)

# 計算cross entropy
cross_entropy_A = -np.sum(P * np.log(Q_A))
cross_entropy_B = -np.sum(P * np.log(Q_B))
cross_entropy_C = -np.sum(P * np.log(Q_C))

# 輸出結果
print(f"模型 A（預測正確一）")
print(f"  KL Divergence: {kl_A:.4f}, cross entropy: {cross_entropy_A:.4f}\n 兩者差距: {cross_entropy_A - kl_A:.4f}\n 兩者比值: {cross_entropy_A / kl_A:.4f}\n")

print(f"模型 B（預測正確二）")
print(f"  KL Divergence: {kl_B:.4f}, cross entropy: {cross_entropy_B:.4f}\n 兩者差距: {cross_entropy_B - kl_B:.4f}\n 兩者比值: {cross_entropy_B / kl_B:.4f}\n")

print(f"模型 C（預測錯誤）")
print(f"  KL Divergence: {kl_C:.4f}, cross entropy: {cross_entropy_C:.4f}\n 兩者差距: {cross_entropy_C - kl_C:.4f}\n 兩者比值: {cross_entropy_C / kl_C:.4f}\n")

print(f"模型 A 的 KL Divergence: {kl_A:.4f}, 模型 B 的 KL Divergence: {kl_B:.4f}\n 兩者比值: {kl_A / kl_B:.4f}\n")
print(f"模型 A 的 cross entropy: {cross_entropy_A:.4f}, 模型 B 的 cross entropy: {cross_entropy_B:.4f}\n 兩者比值: {cross_entropy_A / cross_entropy_B:.4f}\n")
```
