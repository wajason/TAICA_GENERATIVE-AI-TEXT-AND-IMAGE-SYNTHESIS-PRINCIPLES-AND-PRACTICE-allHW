```python
import requests             # requests 的HTTP 請求方法
from bs4 import BeautifulSoup     # 爬蟲工具
import time               # 時間存取與轉換

# 模擬正常用戶行為，提高請求的成功率，還能繞過某些反爬蟲措施。
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# 定義空的articles新字典來儲存後續資料
articles = []
start_page = 7002  # 從第7002頁開始爬蟲
end_page = start_page - 270  # 總共抓270頁

for page in range(start_page, end_page, -1):  # 迴圈抓270頁的內容
    url = f"https://www.ptt.cc/bbs/Food/index{page}.html" # ptt food看板
    response = requests.get(url, headers=headers)    # 發送 GET 請求
    soup = BeautifulSoup(response.text, "html.parser")  # 使用 BeautifulSoup 對獲取的 HTML 內容進行解析

    # 抓每頁的文章連結
    for entry in soup.select(".title a"):
        title = entry.text
        link = "https://www.ptt.cc" + entry["href"]

        # 進入文章抓內文
        try:
            article_response = requests.get(link, headers=headers)
            article_soup = BeautifulSoup(article_response.text, "html.parser")
            main_content = article_soup.select_one("#main-content")
            content = main_content.text.split("※ 發信站")[0] if main_content else "無內文"  # 取正文部分
            articles.append({"title": title, "content": content})
            print(f"已抓取: {title}")
        except Exception as e:
            print(f"抓取失敗: {link}, 錯誤: {e}")

        time.sleep(1)  # 每篇休息1秒(避免被封)

    print(f"已完成第 {page} 頁")
    time.sleep(2)  # 每頁休息2秒(避免被封)

# 存成txt，取前5000篇
with open("ptt_food.txt", "w", encoding="utf-8") as f: # w, 打開一個文件只用於寫入
    for i, article in enumerate(articles[:5000], 1):  # 限制5000篇
        f.write(f"標題 {i}: {article['title']}\n內文 {i}: {article['content']}\n\n")

print("抓取完成~~")
```

```python
from langchain_community.document_loaders import TextLoader, PyPDFLoader, UnstructuredWordDocumentLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
```

### 3. 依 e5 建議加入

自訂支援 E5 的 embedding 模型（加上 "passage:" / "query:" 前綴）

```python
from langchain.embeddings import HuggingFaceEmbeddings # 用Hugging Face模型（如E5）生成文字向量

# 繼承HuggingFaceEmbeddings，使得能進一步修改
class CustomE5Embedding(HuggingFaceEmbeddings):
    def embed_documents(self, texts):
        texts = [f"passage: {t}" for t in texts]  # 要求文件加上"passage:"前綴
        return super().embed_documents(texts)

    def embed_query(self, text):
        return super().embed_query(f"query: {text}")  # 要求文件加上"query: "前綴
```

### 4. 載入文件

```python
folder_path = upload_dir
documents = []  # 初始化一個空列表用來儲存資料
# 逐一處理每個文件
for file in os.listdir(folder_path):
    # 把資料夾路徑和檔案名組合起來
    path = os.path.join(folder_path, file)
    if file.endswith(".txt"):
        loader = TextLoader(path)
    elif file.endswith(".pdf"):
        loader = PyPDFLoader(path)
    elif file.endswith(".docx"):
        loader = UnstructuredWordDocumentLoader(path)
    else:
        continue
    # 載入資料準備分割
    documents.extend(loader.load())
```

### 5. 建立向量資料庫

```python
# 文字分割器
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200) # 每段最多1000字，彼此可以重複200字
split_docs = splitter.split_documents(documents) # 切割好的資料存進split_docs
```

```python
embedding_model = CustomE5Embedding(model_name="intfloat/multilingual-e5-large") # 創建E5-large模型
vectorstore = FAISS.from_documents(split_docs, embedding_model) # 建立向量資料庫
```

### 6. 儲存向量資料庫

```python
vectorstore.save_local("faiss_db")
```
